```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.array([2,4,6,8,10])
y=np.array([3,6,9,10,15])
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'o')
plt.show()
```

```python
plt.plot(x,y,'*:b')    #marker,line,color
plt.show()

# - Solid line
# : dotted line
# -- Dashed line
# -. dashed dotted line
```

```python
plt.plot(x,marker='*',ms=15,mec='r',mfc='#3EB41A')    #marker,line,color
plt.show()
```

```python
plt.plot(x,marker='+',ms=15,mec='r',mfc='#F4D03F',linestyle=':',color='#2E4053',linewidth=5)
plt.plot(y,marker='*',ms=15,mec='r',mfc='#F4D03F',linestyle=':',color='g',linewidth=3)#marker,line,color

f1={'family':'serif','color':'blue','size':20}

plt.xlabel('X Array')
plt.ylabel('Y Array')
plt.title('X vs Y array',fontdict=f1,loc='left')
plt.grid(axis='y')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x)
plt.title('x')

plt.subplot(1,2,2)
plt.plot(y)
plt.title('Y')

plt.suptitle('Relationship between X and Y in different plots')

plt.show()
```

```python
x=np.array(["A","B","C","D","E"])
y=np.array([10,5,20,13,9])

plt.bar(x,y)
plt.show()
```
